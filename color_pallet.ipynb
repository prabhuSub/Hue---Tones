```python
import PIL
from PIL import Image
from matplotlib import pyplot as plt

im = Image.open('./images/Van Gogh starry_night_full_1889.jpg')  
w, h = im.size  
colors = im.getcolors(w*h)

def hexencode(rgb):
    r=rgb[0]
    g=rgb[1]
    b=rgb[2]
    return '#%02x%02x%02x' % (r,g,b)

for idx, c in enumerate(colors):
     plt.bar(idx, c[0], color=hexencode(c[1]),edgecolor=hexencode(c[1]))

plt.show()
```

### Extracting colors from the image and converting them to HEX values. Trying to plot Histogram based on their frequency in the images

```python
from PIL import Image
im = Image.open('./images/Van Gogh starry_night_full_1889.jpg')
from collections import defaultdict
by_color = defaultdict(int)
for pixel in im.getdata():
    by_color[pixel] += 1
# by_color
sort_colors = sorted(by_color.items(), key=lambda x: x[1], reverse=True)
diction_color = {}
for section in sort_colors:
    diction_color[section[0]] = section[1]
# diction_color
print(len(diction_color.keys()))
print(len(diction_color.values()))
keys = list(diction_color.keys())
values = list(diction_color.values())
hex_value_color = []
for k in keys:
    hex_value_color.append('#%02x%02x%02x' % k)
# hex_value_color
```
